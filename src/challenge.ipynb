Para optimizar el tiempo utilizamos el módulo collections de python para agregar los tweets de manera eficiente. Este enfoque asume que podemos cargar el dataset en memoria.

```python
import json
from collections import Counter, defaultdict

def q1_time(file_path):
    tweets_by_date = defaultdict(list)

    with open(file_path, 'r') as file:
        for line in file:
            tweet = json.loads(line)
            date = datetime.strptime(tweet['date'][:10], "%Y-%m-%d").date()  # La fecha debe estar en el formato 'YYYY-MM-DD'
            user = tweet['user']['username']
            tweets_by_date[date].append(user)

    # Contar el numero de tweets por fecha
    tweet_counts = Counter({date: len(users) for date, users in tweets_by_date.items()})

    # Encontrar las 10 fechas con mayor cantidad de tweets
    top_dates = tweet_counts.most_common(10)

    result = []
    for date, _ in top_dates:
        # Encontrar el usuario con mas tweets para cada fecha
        user_counts = Counter(tweets_by_date[date])
        top_user = user_counts.most_common(1)[0]
        result.append((date, top_user[0]))

    return result
```

**Claramente se puede ver que el metodo que más tiempo consume es strptime y manejar las fechs con datetime**

Podemos reducir drasticamente el tiempo de ejecucion si cambiamos el output de datetime.date a un formato YYYY,MM,DD.

Aun así podemos ganar unos segundos de tiempo eliminando el segundo counter en la anterior función.

```python
import json
from collections import defaultdict, Counter

def q1_time_v2(file_path):
    #Utilizamos un 'nested dictionary' para ahorra la necesidad de un segundo Counter
    tweets_by_date_and_user = defaultdict(lambda: defaultdict(int))
    tweet_counts_by_date = Counter()

    with open(file_path, 'r') as file:
        for line in file:
            tweet = json.loads(line)
            date = datetime.strptime(tweet['date'][:10], "%Y-%m-%d").date()
            user = tweet['user']['username']
            tweets_by_date_and_user[date][user] += 1
            tweet_counts_by_date[date] += 1

    top_dates = tweet_counts_by_date.most_common(10)

    result = []
    for date, _ in top_dates:
      #Aqui ya no necesitamos counter por el diccionario 'tweets_by_date_and_user'
        user_counts = tweets_by_date_and_user[date]
        top_user = max(user_counts, key=user_counts.get)
        result.append((date, top_user))

    return result
```

Para optimizar por memoria, procesamos el archivo linea por linea, de esta manera nunca cargamos el dataset entero en memoria y aprovechamos las lecciones aprendidas anteriormente para disminuir Counters.

```python
import json
from collections import defaultdict

def q1_memory(file_path):
    tweets_by_date = defaultdict(lambda: defaultdict(int))

    with open(file_path, 'r') as file:
        for line in file:
            tweet = json.loads(line)
            date = datetime.strptime(tweet['date'][:10], "%Y-%m-%d").date()
            user = tweet['user']['username']
            tweets_by_date[date][user] += 1

    # Ordenar las fechas por numero de tweets
    sorted_dates = sorted(tweets_by_date.keys(), key=lambda d: sum(tweets_by_date[d].values()), reverse=True)[:10]

    result = []
    for date in sorted_dates:
        # Encontrar el usuario con mas tweets sin usar counter
        top_user = max(tweets_by_date[date], key=tweets_by_date[date].get)
        result.append((date, top_user))

    return result
```

Dado que los intentos previos no disminuyeron la memoria podemos utilizar una estructura de datos un poco más compleja como lo es un heap para encontrar el top 10 sin necesidad de ordenar todo.

```python
import json
from collections import defaultdict
import heapq

def q1_memory_v3(file_path):
    user_tweet_counts = defaultdict(lambda: defaultdict(int))
    total_tweets_by_date = defaultdict(int)

    with open(file_path, 'r') as file:
        for line in file:
            tweet = json.loads(line)
            date = datetime.strptime(tweet['date'][:10], "%Y-%m-%d").date()
            user = tweet['user']['username']
            user_tweet_counts[date][user] += 1
            total_tweets_by_date[date] += 1

    # Utilizamoso un heap para encontrar las 10 fechas con más tweets
    top_dates_heap = [(count, date) for date, count in total_tweets_by_date.items()]
    heapq.heapify(top_dates_heap)
    top_dates = heapq.nlargest(10, top_dates_heap)

    result = []
    for _, date in top_dates:
        top_user = max(user_tweet_counts[date], key=user_tweet_counts[date].get)
        result.append((date, top_user))

    return result
```

Para seguir disminuyendo el tiempo de ejecución utilizaré un patrón de emojis que encontré [aquí](https://gist.github.com/Alex-Just/e86110836f3f93fe7932290526529cd1?permalink_comment_id=3236190#gistcomment-3236190). La diferencia con la libreria emoji es la especificidad de los emojis (ej. color de piel en emoji de las manos 🙏🏻 vs 🙏)

```python
import json
import re
from collections import Counter

# Este es el patron de emojis usualmente utilizado
EMOJI_PATTERN = re.compile(
    "(["
    "\U0001F1E0-\U0001F1FF"  # flags (iOS)
    "\U0001F300-\U0001F5FF"  # symbols & pictographs
    "\U0001F600-\U0001F64F"  # emoticons
    "\U0001F680-\U0001F6FF"  # transport & map symbols
    "\U0001F700-\U0001F77F"  # alchemical symbols
    "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
    "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
    "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
    "\U0001FA00-\U0001FA6F"  # Chess Symbols
    "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
    "\U00002702-\U000027B0"  # Dingbats
    "])"
)

def q2_time_v3(file_path):
    emoji_counter = Counter()

    with open(file_path, 'r') as file:
        for line in file:
            tweet = json.loads(line)
            content = tweet['content']
            # Utilizar regex para encontrar los emojis
            emojis = EMOJI_PATTERN.findall(content)
            # Usar el contador
            emoji_counter.update(emojis)

    top_emojis = emoji_counter.most_common(10)

    return top_emojis
```

Para optimizar aun más la función utilizamos una estructura de datos heap, que nos permite ganar un poco más de rendimiento.

```python
import json
import re
from heapq import nlargest
from collections import defaultdict

def q2_time_v4(file_path):
    emoji_counts = defaultdict(int)  # Utilizar un diccionario para contar

    with open(file_path, 'r') as file:
        for line in file:
            tweet = json.loads(line)
            content = tweet['content']
            emojis = EMOJI_PATTERN.findall(content)
            for emoji in emojis:
                emoji_counts[emoji] += 1

    # Usamos nlargest porque internamente tiene una estructura heap y es mas eficiente
    top_emojis = nlargest(10, emoji_counts, key=emoji_counts.get)

    # Convertimos al output con el formato requerido
    top_emojis_with_counts = [(emoji, emoji_counts[emoji]) for emoji in top_emojis]

    return top_emojis_with_counts
```

```python
import json
from collections import Counter, defaultdict

def q3_time_v2(file_path):
    mentions_count = Counter()
    user_info = defaultdict(str)

    with open(file_path, 'r') as file:
        for line in file:
            tweet = json.loads(line)
            # Siempre va a haber una lista para lidear con errores None
            mentioned_users = tweet.get('mentionedUsers') or []
            for user in mentioned_users:
                user_id = user['id']
                mentions_count[user_id] += 1
                # Actualizar la informacion para usuarios que no se han visto antes
                if not user_info[user_id]:
                    user_info[user_id] = user['username']

    # Usar most common porque usa estructuras heap para ordenar
    top_mentions = mentions_count.most_common(10)
    results = [(user_info[user_id], count) for user_id, count in top_mentions]

    return results
```

```python
import json

def q3_memory(file_path):
    mentions_count = {}
    user_info = {}

    with open(file_path, 'r') as file:
        for line in file:
            tweet = json.loads(line)
            if 'mentionedUsers' in tweet and tweet['mentionedUsers']:
                for user in tweet['mentionedUsers']:
                    user_id = user['id']
                    if user_id in mentions_count:
                        mentions_count[user_id] += 1
                    else:
                        mentions_count[user_id] = 1
                        user_info[user_id] = user['username']

    # Ordenar las menciones
    sorted_mentions = sorted(mentions_count.items(), key=lambda x: x[1], reverse=True)

    results = []
    for user_id, count in sorted_mentions[:10]:
        results.append((user_info[user_id], count))

    return results
```
